## MODEL BUILDING / CREATION :

```python
# spitting the data into x & y
```

```python
# divide the data into input and target
X= dataa.drop('Sales', axis= 1)
y= dataa['Sales']
```

```python
y
```

```python
# Splitting into train & test data 
```

```python
# Divide the x and y into training and testing data 
# 100% data will be there.
```

```python
'''
# For training maximum data will be sent for the model
# to train, (70%, 75%, 80%).

# For testing: minimum data is 20%, 25% or 30% will be 
# sent for testing, (new records) for the model.

# Test data: It is a total unknown record for the model
'''
```

```python
# sklearn is a scientific library which has all the  functionalities of ML

from sklearn.model_selection import train_test_split  # this library is imported to splitting the datas to train & test
 
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.2, random_state= 4)  

# Here testsize  = 0.2 is 20% for testing , random state  = 4 is to saving the particular shuffle of data which is entering in testing side.
```

```python
X.shape # 100% of data
```

```python
X_train.shape # 80% data for training 
```

```python
X_test.shape  # 20% data for test
```

## ALGORITHM

```python
# importing the algorithm
from sklearn.linear_model import LinearRegression # linear regressing model package

# initialize it
model= LinearRegression() # calling required func.  

# train the model
model.fit(X_train, y_train) # Fitting the data into the model
```

```python
# We trained the model
# We need to test the model for the performance
# y pred -> target output 
```

```python
y_pred= model.predict(X_test) # 
```

```python
y_pred # predictions of sales(target variable)
```

```python
# Now we've to compare actual y and predicted y and to find errors;
# errors  = diff b/w actual & predicted
```

```python
y_test
```

```python
# st.ine eqn is y = mx + c
# m-> slope
# c -> intercept
```

```python
#  slopes values are given - (3 cols.) -> m
# m1 x1 + m2 x2 + m3 x3 + c = y
model.coef_
```

```python
# intercept -> c 
model.intercept_
```

## Now error/Loss funcs :

```python
# Error= Actual - Prediction
# MSE
# MAE
# RMSE
# takes 2 parameters y_test, y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse= mean_squared_error(y_test, y_pred)
mse
```

```python
mae= mean_absolute_error(y_test, y_pred)
mae
```

```python
import numpy as np
rmse = np.sqrt(mse)
rmse
```

## R2 score :

```python
r2_score(y_test, y_pred)  # ranges from 0->1 ; higher score indicates a better fit of model (ie) 90% score prediction 
```
